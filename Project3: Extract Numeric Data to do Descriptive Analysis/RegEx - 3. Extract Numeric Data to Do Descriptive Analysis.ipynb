```python
# import library

import pandas as pd
```

```python
# get dataframe

df = pd.read_csv(r'D:\MyWorks\DataForProject\Data For RegEx\Data for Project3.csv')

df
```

### 1.Extract Movies' Title and Year

```python
dm = df['title_year'].str.extract(r'(.+)\s\((\d{4})\)')
dm.columns=['title','year']
dm['year']=dm['year'].astype(int)

dm
```

### 2.Extract Hour and Minute

```python
dt=df['length'].str.extract('(\d+)\shour\(s\)\s(\d+)\sminute\(s\)').astype(float)
dt.columns=['hour','minute']

dt
```

```python
# to get total minute

dt['total_min']=dt['hour']*60+dt['minute']

dt
```

### 3.Extract Budget and Revenue

```python
ds = df['budget_revenue'].str.replace(',','').str.split('/', expand=True).astype(float)
ds.columns=['budget','revenue']

ds
```

```python
# to calculate profit


ds['profit']=ds['revenue']-ds['budget']

ds
```

### 4.Horizontally appand all dataframe to be one dataframe

```python
df=pd.concat([df,dm,dt,ds], axis=1)

df
```

### 5.Do some descriptive analysis from data we have extracted

```python
# find average

df[['total_min','budget','revenue','profit']].mean()
```
