```python
# import labrary

import pandas as pd
import datetime
```

```python
# get dataframe

df = pd.read_excel(r'D:\MyWorks\DataForProject\Data For RegEx\Data for Project2.xlsx',sheet_name='data').set_index('ID')

df
```

```python
# extract date data from 'Description' column

df['Date'] = df['Description'].str.extract('(\d{2,4}[/-]\d{2}[/-]\d{2,4})')

df
```

```python
# create custom function to clean 'Date' data

def cleanData(date):
    date = date.replace('-','/') # replace '-' with '/' to be able split at once
    lst = date.split('/')
    
    if int(lst[2]) > 2000: # to check is the year data occur at the begining or the end
        d = int(lst[0])
        m = int(lst[1])
        y = int(lst[2])
    else:
        d=int(lst[2])
        m = int(lst[1])
        y = int(lst[0])
    
    return datetime.date(y, m, d)
```

```python
# apply custom function to clean 'Date' data

df['Date'] = df['Date'].apply(cleanData)

df
```
