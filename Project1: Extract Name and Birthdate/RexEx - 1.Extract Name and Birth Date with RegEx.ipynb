```python
# import library

import pandas as pd
```

```python
# get dataframe

df = pd.read_excel(r'D:\MyWorks\DataForProject\Data For RegEx\Data for Project1.xlsx',sheet_name='data')

df
```

```python
# extract first name, middle name, and last name from 'President' column by using regular expression

df['First Name'] = df['President'].str.extract('(\w+)')
df['Middle Name'] = df['President'].str.extract('(?:[\w]+\s)(\w+[.]?)(?:\s\w+)')
df['Last Name'] = df['President'].str.extract('([\w+]+$)')

df
```

```python
# use regular expression to clean birthdate data in 'Born' column

df['Born'] = df['Born'].str.extract('(\w{1,3}\s\d{1,2}[,]\s\d{1,4})')

df
```

```python
# extract Year, Days and Dates from 'Age atstart of precidency' column

df['Years of Presidency'] = df['Age atstart of precidency'].str.extract('(\d{1,2})(?:\s)(?:\w+)(?:,)(?:\s)(?:\d{1,3})(?:\s)(?:\w{4})(?:\w.*)')
df['Days of Presidency'] = df['Age atstart of precidency'].str.extract('(?:\d{1,2})(?:\s)(?:\w+)(?:,)(?:\s)(\d{1,3})(?:\s)(?:\w{4})(?:\w.*)')
df['Date of Presidency'] = df['Age atstart of precidency'].str.extract('(?:\d{1,2})(?:\s)(?:\w+)(?:,)(?:\s)(?:\d{1,3})(?:\s)(?:\w{4})(\w.*)')

df
```
