```python
# import library

import pandas as pd
```

```python
# to get dataframe

df=pd.read_excel(r'D:\MyWorks\DataForProject\Data For RegEx\Data for Project4.xlsx', header=None,  names=['text'])

df
```

```python
# first method - by indentifying all Thai character 'ก-๙'

dt = df['text'].str.extract(r'(\w+)\s+\(([ก-๙]+)\)\s+\=\s+([ก-๙]+)')
dt.columns=['english','pronounce','thai']

dt
```

```python
# second method - by specifying a non-numeric characters

dt = df['text'].str.extract(r'(\w+)\s+\((\D+)\)\s+\=\s+(\D+)')
dt.columns=['english','pronounce','thai']

dt
```
